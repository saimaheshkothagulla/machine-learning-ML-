```python
x=file['area']
y=file['price']
xmean=x.mean()
ymean=y.mean()
num,den=0,0
for i in range(len(x)):
    num+=(x[i]-xmean)*(y[i]-ymean)
    den+=(x[i]-xmean)**2
m=num/den
c=ymean-m*xmean
print(m,c)
```

```python
ypred=c+m*x
print(ypred)
```

```python
mp.scatter(file['area'],file['price'])
```

```python
mp.plot(file['area'],file['price'],'ro-')
```

```python
mp.scatter(x,y)
mp.plot([min(x),max(x)],[min(ypred),max(ypred)],color='red')
mp.show()
```

```python
mp.axis([2000,5000,520000,800000])
mp.scatter(x,y)
mp.scatter(x,ypred)
mp.plot([min(x),max(x)],[min(y),max(y)],'g--',label='obtained line')
mp.plot([min(x),max(x)],[min(ypred),max(ypred)],'r--',label='slope line')
mp.legend()
mp.xlabel('x axis')
mp.ylabel('y axis')
mp.show()
```

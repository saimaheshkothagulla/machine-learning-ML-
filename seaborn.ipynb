```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# visualizing static relation ship between the variables of same datasets
by using relplot graphs

```python
a=sns.load_dataset('flights')
```

```python
a.head()
```

```python
sns.relplot(x='passengers',y='month',data=a)
```

```python
sns.relplot(x='passengers',y='month',hue='year',data=a)
```

```python
b=sns.load_dataset('tips')
b.head()
```

```python
sns.relplot(x='time',y='tip',data=b,kind='line')
```

# caterogial plot:if main variable divides into discrete groups
using catplot graphs

we can use the following plots

```python
sns.catplot(x='day',y='total_bill',data=b)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=b)
```

```python
sns.catplot(x='day',y='total_bill',kind='boxen',data=b)
```

# multiplot grid- graphs r ploting side by side using same scale and axex to aid comparison

```python
a=sns.load_dataset('iris')
b=sns.FacetGrid(a,col='species')
b.map(plt.hist,'sepal_length')
```

```python
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```

#changing color pallets for more attractive and delightful by using set function

```python
sns.set(style='darkgrid')
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='dark')
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```
